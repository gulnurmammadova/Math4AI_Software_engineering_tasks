```python
'''Mini layihə
Saytdan məlumat toplayan proqram yazın:
1. Request göndərin
2. HTML götürün
3. Məhsul/kitab adını tapın
4. Qiyməti tapın
5. Nəticəni çap edin'''

import requests
from bs4 import BeautifulSoup


url = "http://books.toscrape.com/"
headers = {"User-Agent": "Mozilla/5.0"} 
response = requests.get(url, headers=headers)
response.encoding = 'utf-8'


if response.status_code == 200:
    html_content = response.text
    soup = BeautifulSoup(html_content, "html.parser")
    

    books = soup.find_all("article", class_="product_pod")
    
    print(f"Səhifədə {len(books)} kitab tapıldı.\n")
    print("-" * 80)

    for book in books:
        title = book.h3.a["title"]
        
        price = book.find("p", class_="price_color").text
        
        print(f"Kitab Adı: {title}")
        print(f"Qiymət   : {price}")
        print("-" * 50)
else:
    print(f"Sayta qoşulmaq mümkün olmadı. Xəta kodu: {response.status_code}")
```

```python
'''1. jsonplaceholder comments API-dən email-ləri 
çıxarın
2. users API-dən name, email, city göstərin'''

import requests

comments_url = "https://jsonplaceholder.typicode.com/comments"
comments_res = requests.get(comments_url)
if comments_res.status_code == 200:
    emails = [comment['email'] for comment in comments_res.json()[:10]] 
    print("--- İlk 10 Komment Email-i ---")
    print(emails)

```

```python
'''
2. users API-dən name, email, city göstərin'''

users_url = "https://jsonplaceholder.typicode.com/users"
users_res = requests.get(users_url)
if users_res.status_code == 200:
    print("\n--- İstifadəçi Məlumatları ---")
    for user in users_res.json():
        print(f"Ad: {user['name']} | Email: {user['email']} | Şəhər: {user['address']['city']}")
```

```python
'''3. books.toscrape.com-dan kitab adı və qiymət 
çıxarın
4. Qiyməti £50-dən aşağı olanları filter edin'''


from bs4 import BeautifulSoup

url = "http://books.toscrape.com/"
response = requests.get(url)
response.encoding = 'utf-8' 

soup = BeautifulSoup(response.text, "html.parser")
books = soup.find_all("article", class_="product_pod")

print("\n--- Qiyməti £50-dən aşağı olan kitablar ---")

for book in books:
    title = book.h3.a["title"]
    price_text = book.find("p", class_="price_color").text # Nümunə: "£51.77"
    
    price_value = float(price_text.replace("£", ""))
    
    if price_value < 50.00:
        print(f"AD: {title}")
        print(f"QİYMƏT: £{price_value}")
        print("-" * 30)
```
