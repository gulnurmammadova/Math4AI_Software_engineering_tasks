```python
'''Employee
 salary __mkr_score name
  salary 1000-2000 mkr salary-40%
  salarfy 2000-5000 mkr salary - 25%
  salarfy 5000 mkr salary-15%'''

class Employee:
    def __init__(self,name, salary, mkr_score):
        self.name=name
        self.salary=salary
        self.__mkr_score=mkr_score

    @property
    def adjusted_salary(self):
        if 1000<=self.__mkr_score<2000:
            return self.salary*0.6
        elif 2000<=self.__mkr_score<5000:
            return self.salary*0.75
        elif self.__mkr_score>=5000:
            return self.salary*0.85
        else:
            return self.salary
    @property
    def mkr_score(self):
        return self.__mkr_score
    
    @mkr_score.setter
    def mkr_score(self, score):
        if score >= 0:
            self.__mkr_score = score
        else:
            print("Score must be non-negative")

    def info(self):
        print(f'{self.name}, based on your mkr score of {self.__mkr_score} your adjusted salary is equal to {self.adjusted_salary}.')

emp1 = Employee('Tom',3000, 6500)        
emp1.info()
```

```python
'''
2.)BankAccount class-ı yaradın.

Tələblər:

owner private olsun
balance private olsun
balance üçün property yazın
balance mənfi ola bilməz
balans birbaşa dəyişdiriləndə (acc.balance = ...) yalnız int və ya float qəbul etsin
əgər balans 10000-dən çox olarsa ekrana "VIP account" yazsın
owner üçün ayrıca property yazın:
ad yalnız hərflərdən ibarət olmalıdır
minimum 3 simvol olmalıdır
deposit(amount) və withdraw(amount) metodları yazın
withdraw zamanı balans çatmırsa custom exception atılsın
'''

class InsufficientFundsError(Exception):
    pass

class BankAccount:
    def __init__(self, owner, balance):
        
        self.owner = owner
        self.__balance = balance

    @property
    def balance(self):
        return self.__balance

    @balance.setter
    def balance(self, value):
        if not isinstance(value, (int, float)):
            raise TypeError('Only numbers are allowed!')
            
        if value < 0:
            raise ValueError('Balance cannot be negative!')
            
        if self.__balance > 10000:
            print('VIP account')
        
        self.__balance = value
      
    @property
    def owner(self):
        return self.__owner

    @owner.setter
    def owner(self, value):
        if len(value) < 3:
            raise ValueError('Minimum of 3 symbols required')
        if not value.isalpha():
            raise ValueError('Name should consist of letters only')
        self.__owner = value

    def deposit(self, amount):
        if not isinstance(amount, (int, float)):
            raise TypeError('Amount must be a number')
        if amount <= 0:
            raise ValueError('Deposit amount must be positive')
        self.balance += amount
        if self.balance > 10000:
            print('VIP account')
        return self.balance
       

    def withdraw(self, amount):
        if not isinstance(amount, (int, float)):
            raise TypeError('Amount must be a number')
        if amount <= 0:
            raise ValueError('Withdraw amount must be positive')
        if amount > self.__balance:
            raise InsufficientFundsError('Not enough balance')
        self.balance -= amount
        return self.balance

if __name__ == "__main__":
    try:
        h1 = BankAccount('Tom', 5300)
        h1.deposit(8000) 
        print(f"Current Balance: {h1.balance}")
        h1.withdraw(1000) 
        print(f"After withdraw: {h1.balance}")
    except (InsufficientFundsError, ValueError, TypeError) as e:
        print(f"Error: {e}")
   
```

```python
'''
3.) Student class-ı yaradın.

Tələblər:

private field-lər:
__name
__age
__grades
average

name property:
yalnız string olsun
age property:
16 ilə 30 arasında olmalıdır
grades property:
yalnız list qəbul etsin
list içindəki bütün qiymətlər 0 ilə 100 arasında olmalıdır
average adlı property yaradın
orta qiyməti qaytarsın
status adlı  property yaradın
average >= 90 → "Excellent"
average >= 75 → "Good"
average >= 60 → "Normal"
əks halda "Fail"
'''


class Student:
    def __init__(self, name, age, grades):
        self.name = name
        self.age = age
        self.grades = grades

    @property
    def name(self):
        return self.__name
    
    @name.setter
    def name(self, value):
        if not isinstance(value, str):
            raise TypeError('Name should be a string')
        self.__name = value

    @property
    def age(self):
        return self.__age

    @age.setter
    def age(self, value):
        if not 16 <= value <= 30:
            raise ValueError('Age must be between 16 and 30')
        self.__age = value

    @property
    def grades(self):
        return self.__grades

    @grades.setter
    def grades(self, value):
        if not isinstance(value, list):
            raise TypeError('Grades must be a list')
           
        
        for i in value:
            if not (0 <= i <= 100):
                raise ValueError(f'Invalid grade found: {i}. Must be between 0 and 100.')
                
        
        self.__grades = value

    @property
    def average(self):
        if not self.__grades:
            return 0
        return sum(self.__grades) / len(self.__grades)

    @property
    def status(self):
        avg = self.average 
        if avg >= 90:
            return "Excellent"
        elif avg >= 75:
            return "Good"
        elif avg >= 60:
            return "Normal"
        else:
            return "Fail"

if __name__ == "__main__":
    try:
        s1 = Student("Ayla", 20, [35, 80, 90])
        print(f"Name: {s1.name}")
        print(f"Average: {s1.average:.2f}")
        print(f"Status: {s1.status}")
    except (ValueError, TypeError) as e:
        print(f"Error: {e}")
```
